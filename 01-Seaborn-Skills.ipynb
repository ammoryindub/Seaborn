```python
# Import libraries


import numpy as np 
import matplotlib.pyplot as plt
%matplotlib inline 

import seaborn as sns
```

```python
# we will import a dataset that is already installed into Seaborn named 'planets'

df = sns.load_dataset('planets')
```

```python
# now we can check our dataframe we imported from Seaborn datasets

df 
```

```python
# we will apply dropna() method to cleans our data before we constinue

ds = df.dropna()
```

```python
ds
```

```python
ds.info()
```

```python
sns.jointplot(data=ds, x='mass', y='orbital_period')

plt.show()
```

```python
sns.distplot(ds['distance'])

plt.show()
```

```python
# lets add more arguments to our distplot

sns.distplot(ds['distance'], kde=False, color='green', bins=40)
```

```python
sns.boxplot (x='number', y='mass', data=ds, palette='rainbow')

plt.show()
```

```python
sns.swarmplot (x='number', y='mass', data=ds, palette='rainbow')

plt.show()
```

Will demonstrateour dataset using different ways of data visualization

#### Distribtional Plots 

```python
# Histogram plotting


# Set the style
sns.set_style('whitegrid')

# Plot a histogram of the 'mass' column with custom color
sns.histplot(data=ds, x='mass', color='steelblue')

# Add additional customization if needed

# Display the plot
plt.show()
```

```python
# KDE Plot


# Set the style
sns.set_style('darkgrid')

# Plot a KDE plot of the 'mass' column with custom color
sns.kdeplot(data=ds, x='mass', color='green')

# Add additional customization if needed

# Display the plot
plt.show()
```

```python
# violin plot

# Set the style
sns.set_style('ticks')

# Plot a box plot of the 'distance' column with custom color
sns.violinplot(data=ds, y='distance', color='purple')

# Add additional customization if needed

# Display the plot
plt.show()
```

#### Categorical Plots 

```python
# Bar plot

sns.set_style('whitegrid')

# Plot a bar plot of the 'method' column with custom color
sns.countplot(data=ds, x='method', color='steelblue')

# Add additional customization if needed

# Display the plot
plt.show()
```

```python
# Point plot

# Set the style
sns.set_style('darkgrid')

# Plot a point plot of the 'method' column with custom color
sns.pointplot(data=ds, x='method', y='distance', color='green')

# Add additional customization if needed

# Display the plot
plt.show()
```

```python
# Swarm plot

# Set the style
sns.set_style('ticks')

# Plot a swarm plot of the 'year' column with custom color
sns.swarmplot(data=ds, y='year', color='orange')

# Add additional customization if needed

# Display the plot
plt.show()
```

```python
# Boxen plot

# Set the style
sns.set_style('dark')

# Plot a boxen plot of the 'method' column with custom color
sns.boxenplot(data=ds, x='method', y='orbital_period', color='purple')

# Add additional customization if needed

# Display the plot
plt.show()
```

#### Matrix/Grid Plots 

```python
# Heatmap 

# Set the style
sns.set_style('whitegrid')

# Create a correlation matrix of the data
correlation_matrix = ds.corr()

# Plot a heatmap of the correlation matrix with custom color
sns.heatmap(correlation_matrix, cmap='coolwarm')

# Add additional customization if needed

# Display the plot
plt.show()
```

```python
# Pairplot

# Set the style
sns.set_style('darkgrid')

# Plot a pairplot of selected columns with custom color
columns_to_plot = ['mass', 'distance', 'year']
sns.pairplot(data=ds[columns_to_plot], diag_kind='kde', palette='Set2')

# Add additional customization if needed

# Display the plot
plt.show()
```

```python
# Clustermap

# Set the style
sns.set_style('white')

# Create a subset of data for clustering
subset_data = ds[['mass', 'distance', 'year']]

# Plot a clustermap of the subset data with custom color
sns.clustermap(subset_data, cmap='viridis')

# Add additional customization if needed

# Display the plot
plt.show()
```

#### Regression plot

```python
# Set the style
sns.set_style('whitegrid')

# Plot a regression plot between 'mass' and 'distance' columns with custom color
sns.regplot(data=ds, x='mass', y='distance', color='steelblue')

# Add additional customization if needed

# Display the plot

plt.show ()
```
